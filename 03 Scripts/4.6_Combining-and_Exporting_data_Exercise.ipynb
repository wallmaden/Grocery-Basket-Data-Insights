## 02. Merging Data

```python
data3 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'days_purchased_on':[0, 10, 4, 1]}
```

```python
df_2 = pd.DataFrame(data3,index=[0, 1, 2, 3]) # now we have 2 different df of different sizes
```

```python
# merge 2 df together
df_merged = df.merge(df_2, on = 'customer_id')
```

```python
df_merged
```

##### Be careful with the syntax here, as it’s different from that of the pd.concat() function. 
##### Instead, it takes one dataframe (df) and combines it with another dataframe, which you include within the parentheses (df_2). 
##### The addition of the on = customer_id column tells pandas that the “customer_id” column is the common column between the two. 
##### The result is a wide-format dataframe with all the information from both dataframes and, most importantly, no rows with missing values. 

```python
# Specify columns as key to create a 'single' month column
df_merged = df.merge(df_2, on = ['customer_id', 'month'])
```

```python
df_merged
```

```python
# Check for a full match
df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
df_merged
```

```python
df_merged['_merge'].value_counts()
```

```python
# Test merge without overwriting, must include both df's as this is a creating a new df:
pd.merge(df,df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
# the 'how' argument, how you want the dataframes to be merged, and it can take the values left, right, inner (default), or outer, e.g. inner join
df.merge(df_2, on = ['customer_id', 'month'], how = 'inner')
```
