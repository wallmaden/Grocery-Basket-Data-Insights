### Step 6

#### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Calculate the mean price for each user and add it to the DataFrame
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check the new 'average_spend' column
ords_prods_merge.head()
```

```python
# Assign a 'Low_spender' tag to each user lower than 10 mean

ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'
```

```python
# Assign a 'High_spender' tag to each user higher than 10 mean

ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
ords_prods_merge.head() # spender_flag column now in the df
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

### Step 7

#### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# First create the median in the “days_since_last_order” column...call it 'order_median_gap'

ords_prods_merge['order_median_gap'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
# Check that the column was created correctly
ords_prods_merge.head()
```

```python
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

ords_prods_merge.loc[ords_prods_merge['order_median_gap'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

ords_prods_merge.loc[(ords_prods_merge['order_median_gap'] > 10) & (ords_prods_merge['order_median_gap'] <= 20), 'customer_frequency'] = 'Regular customer'
```

```python
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

ords_prods_merge.loc[ords_prods_merge['order_median_gap'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge.head() # new column now listed on df
```

```python
# Check value counts for new 'customer frequency' column
ords_prods_merge['customer_frequency'].value_counts(dropna = False)
```

##### NB. NB. The total count for the newly created 'customer_frequency' column matches the total count from the "busiest_day" column previously. This consistency confirms that no data has been lost or misclassified during the column creation process, and both columns accurately reflect the dataset.

##### The majority of customers are classified as "Frequent customers" (21.56M), followed by "Regular customers" (7.21M) and "Non-frequent customers" (3.63M), with only 5 entries having missing frequency data.
